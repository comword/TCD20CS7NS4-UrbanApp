```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu, sigma = 0.05, 3
s = np.random.normal(mu, sigma, 420)
```

```python
s = abs(s)
```

```python
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

```python
np.savetxt("rainfall.csv", s)
```
